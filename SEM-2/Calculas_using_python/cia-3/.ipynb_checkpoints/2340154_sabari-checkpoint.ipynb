3) Write a program to factorize the numerator of $f(x)=\frac{4x^3-3x-1}{4x^2+4x+1}$ and simplify $f(x)$.

```python
from sympy import Symbol, diff

x = Symbol('x')
eqn = (4*x**3 -3*x -1)/ (4*x**2 +4*x+1)
y = diff(eqn, x)
print(y)
```

1) Identify the error in the following snippet.

```python
from numpy import *
from scipy.integrate import odeint
import matplotlib.pyplot as plt
def logistic_growth(P,t,r,K):
    dPdt=r*P*(1-P/K)
    return dPdt
P0=100 #Initial population size
r=0.1
K=1000
t= linspace(0,100,1000) #Time points for which to sole the ODE
solution=odeint(logistic_growth,P0,t,args=(r,K))
plt.figure(figsize=(10,6))
plt.plot(t,solution,label="Population Size", color='r')
plt.xlabel("Time")
plt.ylabel("Population Size")
plt.title("Logistic Growth")
plt.legend()
plt.grid(True)
plt.show()
```

2) Write a Python program to obtain the Taylor's polynomial of $\cos(x)$ centred at $x=1$ and the Maclaurin polynomial of $\cos(x)$. The degree of both the polynomials is $2$.

```python

import numpy as np
import sympy as sp

x = Symbol('x')

cos_x = cos(x)
taylor_poly = cos_x.series(x, 1, 2)
print(taylor_poly)

mac_poly = cos_x.series(x,1,2)
print(mac_poly)

```
